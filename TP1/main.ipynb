```python
import matplotlib.pyplot as plt

t = np.arange(-10, 10, 0.01)

f = np.sin(2 * np.pi * t) + np.sin(np.pi * t)

plt.plot(t, f, 'r')
plt.show()
```

```python
from mpl_toolkits import mplot3d

def f(x,y):
    return np.sin(x*y) + np.cos(x)

x = np.linspace(-4,4,50)
y = np.linspace(-4,4,50)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis', edgecolor='none')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
import sympy as sp

x=sp.symbols('x')         
y=sp.symbols('y')

fxy = sp.sin(x*y)+sp.cos(x)
fn = sp.lambdify((x,y), fxy, "numpy")

xn = np.linspace(-4,4,50)
yn = np.linspace(-4,4,50)

X, Y = np.meshgrid(xn, yn)

Z = fn(X, Y)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis', edgecolor='none')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python

t = np.arange(0,11,1)
y = np.array([0.0, 0.7, 2.4, 3.1, 4.2, 4.8, 5.7, 5.9, 6.2, 6.4, 6.3])

p = np.polyfit(t, y, 2)
ya = np.polyval(p, t)

plt.plot(t, y, 'ok')
plt.plot(t, ya, '--r')
plt.xlabel("t")
plt.ylabel("y")
plt.title("idk")
plt.show()
```

```python
def factorial(n):
    soma = 1
    for i in range(n, 1, -1):
        soma *= i
    print(soma)

factorial(5)
```

```python
def fib(n):
    fib = []
    if n == 1:
        fib.append(0)
    else:
        fib.append(0)
        fib.append(1)
        for i in range(2, n):
            fib.append(fib[i-1]+fib[i-2])
    print(fib)

fib(7)
```
