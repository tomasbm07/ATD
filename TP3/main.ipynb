```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
import scipy as sp
```

```python
with open("lisbon_temp_fmt.txt", "r") as f:
    lines = f.readlines()

values = [0 for i in range(len(lines))]
for i in range(len(lines)):
    values[i] = float(lines[i])

#print(values)
```

```python
x = np.linspace(1980, 2018, len(values))
plt.plot(x, values, '-.')
plt.xlabel("Ano")
plt.ylabel("Temperatura(ºC)")
plt.show()
```
