### Pergunta 6 (2)
Completar o código em *Matlab*/*Python* que permite visualizar, no mesmo gráfico, a representação do 
sinal de tempo contínuo $x(t) = 2sin(3t)^2$ para $t \in [-2\pi, 2\pi]s$, considerando $t$ com 800 elementos e 
um traçado com traço-ponto a verde, e o correspondente sinal de tempo discreto $x[n]$, considerando 
um período de amostragem (passo) $T_{s}$ = 0.05s, num intervalo para n correspondente a $t \in [-2\pi, 2\pi]s$
e a representação apenas das amostras com a marca '+' a amarelo.

```python
t = np.linspace(-2*np.pi, 2*np.pi, 800) # vetor de tempo continuo
xt = 2*np.sin(3*t)**2 # sinal de tempo continuo
Ts = 0.05 # periodo de amostragem
n = np.arange(-2*np.pi, 2*np.pi, Ts) # vetor de tempo discreto
xn = 2*np.sin(3*n)**2 # sinal de tempo discreto
plt.figure()
plt.plot(t, xt, '-.g', n, xn, '+y') # representaçao de x(t) e x[n]
plt.xlabel('Tempo [s]')
plt.ylabel('Sinais $x(t)$ e $x[n]$')
plt.title('Representaçao de $x(t)$ e de $x[n]$ com $T_{s} = 0.05Hz$')
plt.plot()
```

* $x(t) = 2 + 4cos(2t)$ () par  () ímpar  (x) nenhum
* $x(t) = 4sin(3t)cos(2t)$ () par  (x)ímpar  () nenhum

* $x(t) = x(-t) \rightarrow$ par
* $-x(t) = x(-t) \rightarrow$ ímpar
* Sempre que tem uma constante a somar, nunca pode ser ímpar. Ex: $x(t) = \overbrace{2}^{constante} + 4sin(3t)$

NOTA: pode ser pedido as expressões componente par e/ou ímpar de uma funcão (semi-soma - par ; semi-diferença - ímpar)


```python
t = np.linspace(-2*np.pi, 2*np.pi, 1000)
xt = 2+ 4*np.sin(3*t)

plt.plot(t, xt, 'k', label = 'x(t)')
plt.plot(-t, xt, '+r', label = 'x(-t)')
plt.plot(t, -xt, '--b', label = '-x(t)')
plt.legend()
plt.show()
```
